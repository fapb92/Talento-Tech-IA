**Clasificación de texto utilizando Naive Bayes de NTLK**

```python
import random
```

```python
label = {
    "n": "negative",
    "p":"positive"
}

data = [
    ("I love this movie", label["p"]),  # Positive label
    ("This movie is terrible", label["n"]),  # Negative label
    ("This movie is great", label["p"]),  # Positive label
    ("I dislike this movie", label["n"]),  # Negative label
    ("This film is amazing", label["p"]),  # Positive label
    ("I can't stand watching this movie", label["p"]),  # Positive label
    ("The acting in this movie is phenomenal", label["p"]),  # Positive label
    ("I regret wasting my time on this film", label["n"]),  # Negative label
    ("I thoroughly enjoyed this movie", label["p"]),  # Positive label
    ("This movie lacks depth and substance", label["n"]),  # Negative label
    ("The plot of this movie was captivating", label["p"]),  # Positive label
    ("I found the characters in this film to be very engaging", label["p"]),  # Positive label
    ("The special effects in this movie were impressive", label["p"]),  # Positive label
    ("The storyline was predictable and unoriginal", label["n"]),  # Negative label
    ("I was disappointed by the lack of character development", label["n"]),  # Negative label
    ("The cinematography in this film was stunning", label["p"]),  # Positive label
    ("The dialogue felt forced and unnatural", label["n"]),  # Negative label
    ("The pacing of the movie was too slow for my liking", label["n"]),  # Negative label
    ("I was pleasantly surprised by how much I enjoyed this film", label["p"]),  # Positive label
    ("The ending left me feeling unsatisfied and confused", label["n"]),  # Negative label
    ("This movie exceeded my expectations", label["p"]),  # Positive label
    ("The performance by the actors were lackluster", label["n"]),  # Negative label
]
```

```python
#procesamiento de datos: tokenización y extraccíon de caracteristicas
def preprocess(text):
  tokens = nltk.word_tokenize(text)
  return {word: True for word in tokens}
```

***Aplicación del preprocesamiento**

```python
featuresets = [(preprocess(text), label) for (text, label) in data]
```

***División de datos**

```python
# Se dividen los datos en conjuntos de entrenamiento y prueba
train_set, test_set = featuresets[:16], featuresets[16:]
```

**Entrenamiento de clasificador**

```python
# Se entrena un clasificador utilizando Naive Bayes
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

**Evaluación de clasificador**

```python
accuracy = nltk.classify.accuracy(classifier, test_set)
print("Accuracy", accuracy)
```

**Clasificación de un nuevo texto**

```python
new_text = "This movie is a trash"
new_text_features = preprocess(new_text)
predicted_label = classifier.classify(new_text_features)
print("Predicted label:",predicted_label)
```
