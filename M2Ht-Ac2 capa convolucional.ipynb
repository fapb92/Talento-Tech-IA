```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
# imagen
image = cv2.imread('imagen.jpg')

# kernel convolución
kernel = np.array([[0,-1,0],
                   [-1,5,-1],
                   [0,-1,0]])

# Se aplica convolución
convolved_image = cv2.filter2D(image, -1, kernel)

# Se muestran las imagen original y convolucionada
print('original')
cv2_imshow(image)
print('Convolved Image')
cv2_imshow(convolved_image)
```

```python
# se define un nuevo filtro
kernel2 = np.array([[-1,0,1],
                    [-2,0,2],
                    [-1,0,1]])

# Se aplica convolución
convolved_image = cv2.filter2D(image, -1, kernel2)

# Se muestran las imagen original y convolucionada
print('original')
cv2_imshow(image)
print('Convolved Image')
cv2_imshow(convolved_image)
```
