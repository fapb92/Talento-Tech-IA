```python
import numpy as np
import keras as kr
```

```python
model = kr.Sequential([kr.layers.Dense(units=1,input_shape=[1])])
```

```python
model.compile(
    optimizer='sgd',
    loss='mean_squared_error'
)
```

```python
X = np.array([1,2,3,4,5],dtype=float)
Y = np.array([11,21,31,41,51],dtype=float)
```

```python
model.fit(X,Y,epochs=100)
```

```python
print(model.predict([10.0]))
```
