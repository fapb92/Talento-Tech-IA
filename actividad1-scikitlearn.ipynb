**Ejemplo de clasificación con *MLPClassifier***

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
X, Y = iris.data, iris.target
```

```python
type(iris)
iris.keys()
iris['data']
iris['target']
iris['target_names']
iris['DESCR']
iris['feature_names']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
mlp_clf = MLPClassifier(hidden_layer_sizes=(100,),
                        activation='relu',
                        solver='adam',
                        max_iter=100,
                        random_state=42,
                        verbose=True)
```

```python
mlp_clf.fit(X_train_scaled,Y_train)
```

```python
Y_pred = mlp_clf.predict(X_test)
print(Y_test,Y_pred,sep='\n\n')
```

```python
accuracy = accuracy_score(Y_test, Y_pred)
print("Precisión del modelo:", accuracy)
```

**Ejemplo de regresión con *MLPRegressor***

```python
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
housing = fetch_california_housing()
X, Y = housing.data, housing.target
```

```python
type(housing)
housing.keys()
housing['data']
housing['target']
housing['target_names']
housing['DESCR']
housing['feature_names']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
mlp_reg = MLPRegressor(hidden_layer_sizes=(100,),
                        activation='relu',
                        solver='adam',
                        max_iter=100,
                        random_state=42,
                        verbose=True)
```

```python
mlp_reg.fit(X_train_scaled, Y_train)
```

```python
Y_pred = mlp_reg.predict(X_test)
print(Y_test,Y_pred,sep='\n\n')
```

```python
import numpy as np
def mse_func(test,pred):
  return np.array([(t-p)**2 for t, p in zip(test,pred)]).mean()
```

```python
mse = mean_squared_error(Y_test, Y_pred)
print("Error cuadrático medio del modelo:",mse)
```

```python
mse_func(Y_test,Y_pred) - mse
```
